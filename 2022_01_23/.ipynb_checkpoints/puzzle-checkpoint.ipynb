## Time to play the puzzle

This week's puzzle: Take the last two letters of a state's capital city and the first two letters of the state. Rearrange them to name an activity that that state is associated with.


```python
import pandas as pd
import csv
import itertools
from PyDictionary import PyDictionary
```

```python
# Read in the data

df = pd.read_csv('capitals.csv', header=0, delimiter=';')
df.head()
```

```python
states = df['State']
states.head()
```

```python
cities = df['City']
cities.head()
```

```python
#get the first two letters of state and last two letters of city. seem to have a leading space on states (whoops) hence the use of 1 & 2 rather than 0 & 1
letters = []
for i in range(len(states)):
    letter1 = states[i][1]
    letter2 = states[i][2]
    letter3 = cities[i][-1]
    letter4 = cities[i][-2]
    let = letter1.lower() + letter2 + letter3 + letter4
    letters.append(let)
letters
```

```python
#Get the different combinations of the letters

df_combos = pd.DataFrame()
for i in range(len(letters)):
    combos = []
    st = letters[i]
    per = itertools.permutations(st)
    for val in per:
        word = (val[0]+val[1]+val[2]+val[3])
        # print(*val)
        combos.append(word)
    # print(combos)
    ser = pd.Series(data=combos, name=states[i])
    df_combos[states[i]] = ser
df_combos
```

```python
#export to csv
df_combos.to_csv('answers.csv')
```

```python
# NLTK dictionary

import nltk
from nltk.corpus import words
```

```python
# Convert words from corpus to a set to make it faster (according to stackoverflow?)
wordset = set(words.words())
```

```python
def test(str):
    wordcheck = False
    if str in wordset:
        wordcheck = True
    return wordcheck

test('hello')
# test('helo')
```
